```python
#importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
#importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#data collection
heart_data = pd.read_csv('/content/heart.csv')
```

```python
#splitting target and features
X = heart_data.drop(columns ='target',axis =1)
Y = heart_data['target']
```

```python
print(X)
print(Y)
```

```python
#1st TRAIN_TEST_SPLIT
#splitting the data into training data and testing data
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size=0.1,stratify=Y,random_state=3)
```

```python
# creating a list to add all the models because training and testing the models individually takes more time so we are going to use a loop function

models = [LogisticRegression(max_iter=1000),SVC(kernel= 'linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
#CREATING A function
def compare_models_train_test():

  for model in models:
    model.fit(X_train,Y_train)
    model_prediction = model.predict(X_test)
    accuracy = accuracy_score(model_prediction,Y_test)
    print('acccuracy of the',model,'=' ,accuracy)
```

```python
#calling the function
compare_models_train_test()
```

#**CROSS VALIDATION COMES IN**

```python
#logistic regression
cv_score_lr = cross_val_score(LogisticRegression(max_iter=1000),X,Y,cv=5)
print(cv_score_lr)
mean_accuracy_lr = sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_lr = mean_accuracy_lr*100
mean_accuracy_lr = round(mean_accuracy_lr,2)
print(mean_accuracy_lr)
```

SINGLE CALL

```python
models = [LogisticRegression(max_iter=1000),SVC(kernel= 'linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
#defining a function
def compare_models_cross_validation():
  for model in models:
    cv_score = cross_val_score(model,X,Y, cv=5)
    accuracy = sum(cv_score)/len(cv_score)
    accuracy = accuracy*100
    accuracy = round(accuracy,2)
    print('accuracy of the',model,':' ,accuracy)
    print('______________________________________________________')
```

```python
compare_models_cross_validation()
```
